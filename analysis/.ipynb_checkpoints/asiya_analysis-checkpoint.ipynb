# モデルの読み込み

```python
model3 = pickle.load(open('asiya_pickle_data/model_com3_dep6_per121.sav', 'rb'))
model4 = pickle.load(open('asiya_pickle_data/model_com4_dep7_per131.sav', 'rb'))
model5 = pickle.load(open('asiya_pickle_data/model_com5_dep8_per122.sav', 'rb'))
model7 = pickle.load(open('asiya_pickle_data/model_com7_dep7_per146.sav', 'rb'))
model13 = pickle.load(open('asiya_pickle_data/model_com13_dep6_per115.sav', 'rb'))
model14= pickle.load(open('asiya_pickle_data/model_com14_dep4_per123.sav', 'rb'))
```

```python
def get_importance(model,model_df):
    #渡したモデルのfeature_importanceを返す。(モデルの予測時に使うdfが必要)
    col_name=model_df.columns
    fti=model.feature_importances_
    
    importances_s=pd.Series(fti,index=col_name)
    #importances_df['importance']=fti
    #importances_df['feature_name']=col_name
    return importances_s

def col_check(df,col1,col2):
    #カラムの順番があっているかどうかの確認
    for index,row in df.iterrows():
        #if row['pred']!=row['model']:
        if row[col1]!=row[col2]:
            print('False\n')
    print('finish')
```

```python
#テスト
model=model3
fti=model.feature_importances_
model_making_df=pd.read_csv('asiya_making_df.csv')
model_making_df=model_making_df.drop(["Unnamed: 0"],axis=1)
importance_s=get_importance(model,model_making_df)

importance_s_sort=importance_s.sort_values(ascending=False)

#importance_plot_ss=importance_df_sort[:50].transpose()
#importance_plot_df.hist()
#importance_df_sort[50:100]
importance_s_sort[:50].plot.bar()
```
